```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from google.colab import files

# Upload the CSV file
uploaded = files.upload()

# Assuming the uploaded file has a .csv extension
csv_file_name = list(uploaded.keys())[0]
```

```python
def read_csv(csv_path):
    data = np.genfromtxt(csv_path, delimiter=',')
    grouped_data = {}

    for row in data:
        label = row[0]  # First column as label
        x, y = row[2], row[3]  # Third and Fourth columns as x and y coordinates

        if label not in grouped_data:
            grouped_data[label] = []

        grouped_data[label].append([x, y])

    for label in grouped_data:
        grouped_data[label] = np.array(grouped_data[label])

    return grouped_data

# Read the uploaded CSV file and group by labels
grouped_data = read_csv(csv_file_name)
```

```python
def plot_raw_data(grouped_data):
    fig, ax = plt.subplots(tight_layout=True, figsize=(8, 8))

    for label, points in grouped_data.items():
        ax.scatter(points[:, 0], points[:, 1], label=f'Label {int(label)}', marker='o')

    ax.set_aspect('equal', 'box')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Raw 2D Points by Label')
    plt.legend()
    plt.grid(True)
    plt.show()

# Visualize the raw data
plot_raw_data(grouped_data)
```

```python
def normalize_data(grouped_data):
    normalized_data = {}
    scaler = MinMaxScaler()  # Using MinMaxScaler for normalization

    for label, points in grouped_data.items():
        normalized_points = scaler.fit_transform(points)
        normalized_data[label] = normalized_points

    return normalized_data

# Normalize the grouped data
normalized_data = normalize_data(grouped_data)
```

```python
def plot_normalized_data(normalized_data):
    fig, ax = plt.subplots(tight_layout=True, figsize=(8, 8))

    for label, points in normalized_data.items():
        ax.scatter(points[:, 0], points[:, 1], label=f'Label {int(label)}', marker='o')

    ax.set_aspect('equal', 'box')
    plt.xlabel('Normalized X')
    plt.ylabel('Normalized Y')
    plt.title('Normalized 2D Points by Label')
    plt.legend()
    plt.grid(True)
    plt.show()

# Visualize the normalized data
plot_normalized_data(normalized_data)
```
