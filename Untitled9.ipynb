```python
# Upload the CSV file
from google.colab import files
import pandas as pd

uploaded = files.upload()

# Read the CSV file
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename, header=None)

# Drop the column that is always zero (column 2)
df_cleaned = df.drop(columns=[1])
df_cleaned.columns = ['Label', 'X', 'Y']

# Display the cleaned dataframe
print(df_cleaned.head())
```

```python
import matplotlib.pyplot as plt

# Visualize the data points (columns X and Y)
plt.figure(figsize=(10, 8))
plt.scatter(df_cleaned['X'], df_cleaned['Y'], s=10, c='blue', label='Data Points')
plt.title('Data Points Visualization')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.legend()
plt.show()
```

```python
# Adjusting DBSCAN parameters
eps_value = 1.0  # Increase this value
min_samples_value = 3  # Decrease this value

# Perform DBSCAN clustering again
db = DBSCAN(eps=eps_value, min_samples=min_samples_value).fit(df_cleaned[['X', 'Y']])
labels = db.labels_

# Number of clusters identified (ignoring noise points)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f'Number of clusters identified: {n_clusters}')

# Update dataframe with labels
df_cleaned['Cluster'] = labels

# Check the number of points in each cluster
print(df_cleaned['Cluster'].value_counts())

# Visualize the new clusters
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

plt.figure(figsize=(10, 8))
for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]  # Black for noise
    class_member_mask = (labels == k)
    xy = df_cleaned[class_member_mask][['X', 'Y']]
    plt.plot(xy['X'], xy['Y'], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6)

plt.title(f'Estimated number of clusters: {n_clusters}')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.show()
```

```python
import cv2
import numpy as np

def detect_rectangles(cluster_points):
    # Define the size of the image (adjust as needed)
    img_size = (500, 500)  # You might need to adjust this based on your data
    img = np.zeros(img_size, dtype=np.uint8)

    # Scale and place points on the image
    for (x, y) in cluster_points:
        if 0 <= int(x) < img_size[1] and 0 <= int(y) < img_size[0]:
            img[int(y), int(x)] = 255  # Place point on the image

    # Find contours
    contours, _ = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    rectangles = []

    for contour in contours:
        # Approximate the contour with a polygon
        epsilon = 0.02 * cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, epsilon, True)

        # Check if the polygon is a rectangle (4 vertices)
        if len(approx) == 4:
            rectangles.append(approx)

    return rectangles

# Example usage:
for cluster_id in df_cleaned['Cluster'].unique():
    cluster_points = df_cleaned[df_cleaned['Cluster'] == cluster_id][['X', 'Y']].values
    rectangles = detect_rectangles(cluster_points)
    print(f'Cluster {cluster_id} contains {len(rectangles)} rectangles.')
```

```python
from skimage.measure import EllipseModel

def detect_ellipses(cluster_points):
    # Convert points to numpy array
    points = np.array(cluster_points, dtype=np.float32)

    if len(points) < 5:
        return None

    # Fit ellipse
    ellipse_model = EllipseModel()
    ellipse_model.estimate(points)

    return ellipse_model.params if ellipse_model.params is not None else None

# Example usage:
for cluster_id in df_cleaned['Cluster'].unique():
    cluster_points = df_cleaned[df_cleaned['Cluster'] == cluster_id][['X', 'Y']].values
    ellipse_params = detect_ellipses(cluster_points)
    if ellipse_params:
        print(f'Cluster {cluster_id} contains an ellipse with parameters: {ellipse_params}')
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def detect_lines(cluster_points):
    """
    Detect lines in a cluster of points using Hough Line Transform.
    """
    # Create a blank image
    img_size = (500, 500)  # Adjust this based on your data
    img = np.zeros(img_size, dtype=np.uint8)

    # Place points on the image
    for (x, y) in cluster_points:
        if 0 <= int(x) < img_size[1] and 0 <= int(y) < img_size[0]:
            img[int(y), int(x)] = 255  # Place point on the image

    # Use Hough Line Transform to detect lines
    lines = cv2.HoughLinesP(img, 1, np.pi / 180, threshold=50, minLineLength=20, maxLineGap=5)
    if lines is not None:
        lines = [line[0] for line in lines]  # Extract lines from the array
    return lines

def draw_lines_on_image(cluster_points, lines):
    """
    Draw detected lines on an image.
    """
    # Create a blank image
    img_size = (500, 500)  # Adjust this based on your data
    img_color = np.zeros((img_size[0], img_size[1], 3), dtype=np.uint8)

    # Place points on the image
    for (x, y) in cluster_points:
        if 0 <= int(x) < img_size[1] and 0 <= int(y) < img_size[0]:
            img_color[int(y), int(x)] = [255, 255, 255]  # White for points

    # Draw lines
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line
            cv2.line(img_color, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green for lines
    else:
        print("No lines detected.")

    plt.figure(figsize=(10, 8))
    plt.imshow(img_color)
    plt.title('Detected Lines')
    plt.show()

# Example usage for a cluster
cluster_id = 0  # Replace with the actual cluster ID you want to analyze
cluster_points = df_cleaned[df_cleaned['Cluster'] == cluster_id][['X', 'Y']].values
lines = detect_lines(cluster_points)
draw_lines_on_image(cluster_points, lines)
```

```python
for cluster_id in df_cleaned['Cluster'].unique():
    cluster_points = df_cleaned[df_cleaned['Cluster'] == cluster_id][['X', 'Y']].values

    # Detect shapes
    rectangles = detect_rectangles(cluster_points)
    ellipse_params = detect_ellipses(cluster_points)
    lines = detect_lines(cluster_points)

    print(f'Cluster {cluster_id} contains:')
    print(f' - {len(rectangles)} rectangles')

    if ellipse_params:
        print(f' - An ellipse with parameters: {ellipse_params}')

    if lines is not None:
        print(f' - {len(lines)} lines')
    else:
        print(f' - No lines detected')
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

def draw_shapes_on_image(cluster_points, rectangles, ellipse_params, lines):
    # Define the size of the image (adjust as needed)
    img_size = (500, 500)  # Adjust this based on your data
    img = np.zeros(img_size, dtype=np.uint8)

    # Place points on the image
    for (x, y) in cluster_points:
        if 0 <= int(x) < img_size[1] and 0 <= int(y) < img_size[0]:
            img[int(y), int(x)] = 255  # Place point on the image

    # Draw rectangles
    img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    for rect in rectangles:
        for i in range(len(rect)):
            pt1 = tuple(rect[i][0])
            pt2 = tuple(rect[(i + 1) % len(rect)][0])
            cv2.line(img_color, pt1, pt2, (0, 255, 0), 2)  # Green for rectangles

    # Draw ellipse
    if ellipse_params:
        center = (int(ellipse_params[0]), int(ellipse_params[1]))
        axes = (int(ellipse_params[2] / 2), int(ellipse_params[3] / 2))
        angle = ellipse_params[4] * 180 / np.pi
        cv2.ellipse(img_color, center, axes, angle, 0, 360, (255, 0, 0), 2)  # Blue for ellipse

    # Draw lines
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line
            cv2.line(img_color, (x1, y1), (x2, y2), (0, 0, 255), 2)  # Red for lines

    plt.figure(figsize=(10, 8))
    plt.imshow(img_color)
    plt.title('Detected Shapes')
    plt.show()

# Example usage for Cluster -1
cluster_id = -1
cluster_points = df_cleaned[df_cleaned['Cluster'] == cluster_id][['X', 'Y']].values
rectangles = detect_rectangles(cluster_points)
ellipse_params = detect_ellipses(cluster_points)
lines = detect_lines(cluster_points)

draw_shapes_on_image(cluster_points, rectangles, ellipse_params, lines)
```
